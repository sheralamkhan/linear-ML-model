# ***Apply linearRegression Model***

```python
from sklearn.linear_model import LinearRegression
```

```python
slr = LinearRegression()
```

```python
y_train.shape
```

```python
singleattribute.shape
```

```python
slr.fit(singleattribute, y_train)
```

```python
model_prediction_train = slr.predict(singleattribute)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
print('mean square error:', mean_squared_error(model_prediction_train, y_train))
print('mean absolute error:', mean_absolute_error(model_prediction_train, y_train))
print('r2 score:', r2_score(model_prediction_train, y_train))
```

# ***Apply Polynomial Regression Model***

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
plr = PolynomialFeatures(degree= 5)

x_trainp = plr.fit_transform(x_train)
x_testp = plr.fit_transform(x_test)
```

```python
plrmodel = LinearRegression()
```

```python
plrmodel.fit(x_trainp, y_train)
```

```python
trainn= plrmodel.predict(x_trainp)
```

```python
print('mean square error:', mean_squared_error(trainn, y_train))
print('mean absolute error:', mean_absolute_error(trainn, y_train))
print('r2 score:', r2_score(trainn, y_train))
```

```python
testn= plrmodel.predict(x_testp)
```

```python
print('mean square error:', mean_squared_error(testn, y_test))
print('mean absolute error:', mean_absolute_error(testn, y_test))
print('r2 score:', r2_score(testn, y_test))
```
